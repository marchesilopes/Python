## Importação

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
```

## Importando a Base de Dados

```python
df_videos=pd.read_excel('videosYT.xlsx',"videos")
display(df_videos)
```

## Tipos de gráficos

Documentação : https://seaborn.pydata.org/api.html#distribution-api

## Criando um gráfico de dispersão (scatterplot):

Documentação : https://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn.scatterplot

```python
fig = sns.scatterplot(data=df_videos, 
                      x= 'Nº de Views',
                      y= 'Nº de Likes',
                      hue= 'Categoria',
                      style= 'Responsável'
                      )
plt.show()
```

## Criando um gráfico de dispersão(relacional - relplot):

Documentação : https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot


```python
fig = sns.relplot(data=df_videos,
                  x= 'Nº de Views',
                  y= 'Nº de Likes',
                  hue= 'Categoria',
                  col= 'Responsável'
                  )
fig.set_titles('O responsável deste gráfico é o {col_name}')
```

## Criando um gráfico de linha (lineplot):

Documentação : https://seaborn.pydata.org/generated/seaborn.lineplot.html#seaborn.lineplot

```python
df_inscritos=pd.read_excel('videosYT.xlsx',"Inscritos")
display(df_inscritos)

fig = sns.lineplot(data=df_inscritos,
                   x= 'Mês/Ano',
                   y= 'Inscritos',
                   color= 'red'
                   )
```

## Criando Histogramas (histplot):

Documentação : https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot


```python
fig = sns.displot(data=df_videos,
                  x='Nº de Views',
                  hue= 'Categoria',                 
                  col= 'Responsável'                  
                  )
```

## Criando o kernel density estimate (KDE) plot (kdeplot):

Documentação : https://seaborn.pydata.org/generated/seaborn.kdeplot.html

```python
fig = sns.kdeplot(data=df_videos,
                  x='Nº de Views'
                  )
```

## Criando empirical cumulative distribution function, ECDF Plot (ecdfplot):

Documentação : https://seaborn.pydata.org/generated/seaborn.ecdfplot.html

```python
fig = sns.ecdfplot(data=df_videos,
                  x='Nº de Views'
                  )
```

## Criando Rug Plot (rugplot):

Documentação : https://seaborn.pydata.org/generated/seaborn.rugplot.html

```python
fig = sns.rugplot(data=df_videos,
                  x='Nº de Views'
                  )
```

## Criando uma regressão linear:

Documentação: https://seaborn.pydata.org/tutorial/regression.html#functions-to-draw-linear-regression-models

```python
fig = sns.regplot ( data= df_videos, x = "Nº de Views", y = "Nº de Likes") 
plt.show(fig)
```

```python
fig = sns.lmplot ( data= df_videos,
                  x = "Nº de Views",
                  y = "Nº de Likes",
                  hue= "Responsável",
                  markers = ['o','x']) 
plt.show(fig)
```
